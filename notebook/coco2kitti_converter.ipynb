```python
import os
```

```python
DATA_PATH = os.getcwd()

kitti_data = 'kitti_data'
if not os.path.isdir(kitti_data):
    os.mkdir(kitti_data)

kitti_data_dir = DATA_PATH + '/kitti_data'
os.chdir(kitti_data_dir)

labels_dir = 'labels'
if not os.path.isdir(labels_dir):
    os.mkdir(labels_dir)

kitti_data_labels = kitti_data_dir + '/labels'
os.chdir(kitti_data_labels)
kitti_data_labels_train = 'train'
kitti_data_labels_val = 'validation'

if not os.path.isdir(kitti_data_labels_train):
    os.mkdir(kitti_data_labels_train)
if not os.path.isdir(kitti_data_labels_val):
    os.mkdir(kitti_data_labels_val)

os.chdir(DATA_PATH)
```

```python
# for each .txt file on tank_data/labels/train and tank_data/labels/validation, access annotID and classID
# convert to kitti format
# save the result in new .txt to kitti_data/labels/train or kitti_data/labels/validation

IDX2LABEL = {0:'dontcare', 1:'Tank', 2:'Tank Cluster', 3:'Floating Head Tank'}

def coco2kitti(mode:str):
  # available mode: train, validation
  target_dir = f'tank_data/labels/{mode}'
  save_dir = f'kitti_data/labels/{mode}/'
  _, _, files = next(os.walk(target_dir))
  for my_file in files:
    file_name = my_file.split('.')[0]
    file_path = target_dir + '/' + my_file
    print(file_path)
    f = open(file_path, "r")
    contents = f.readlines()
    f.close()
    with open(save_dir + file_name + '.txt','w') as label_file:
      for x in contents:
        annot_data = x.split(' ')
        if len(annot_data) == 5:
          clsID = int(annot_data[0].split()[0])
          cls_name = str(IDX2LABEL[clsID])
          bbox_0 = float(annot_data[1].split()[0])
          bbox_1 = float(annot_data[2].split()[0])
          bbox_2 = float(annot_data[3].split()[0])
          bbox_3 = float(annot_data[4].split()[0])

          # Convert COCO bbox coords to Kitti ones
          bbox = [bbox_0, bbox_1, bbox_2 + bbox_0, bbox_3 + bbox_1]
          bbox = [str(b) for b in bbox]
        
          out_str = [cls_name.replace(" ","")
                               + ' ' + ' '.join(['0']*3)
                               + ' ' + ' '.join([b for b in bbox])
                               + ' ' + ' '.join(['0']*8)
                               +'\n']
          label_file.write(out_str[0])
  print('[INFO] success convert data to kitti format')  
  print('[INFO] saving converted data to:', save_dir)  
```

```python
coco2kitti('train')
```

```python
coco2kitti('validation')
```
